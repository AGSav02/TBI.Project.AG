```python
import random

def seq_20(lenght, gc_perc):
    nb_gc = round(gc_perc * lenght)
    nb_au = lenght - nb_gc
    gc_seq = random.choices("GC", k=nb_gc)
    au_seq = random.choices("AU", k=nb_au)
    rna_seq = gc_seq + au_seq
    random.shuffle(rna_seq)
    return "".join(rna_seq)

sequence_2_list = [] 
for i in range(100):
    sequence_2_list.append(seq_20(300, 0.2))

print(len(sequence_2_list))

import RNA
import statistics

md = RNA.md()
mfe_values_2 = []

for i, seq in enumerate(sequence_2_list,1):
    fc = RNA.fold_compound(seq, md)     
    structure, mfe = fc.mfe() 
    mfe_values_2.append(mfe)

mean_mfe_2 = statistics.mean(mfe_values_2)
print(f"Mean minimal free energy: {mean_mfe_2}")


```

```python
with open("random_rna_20.fasta", 'w') as fasta_file:
    for i, seq in enumerate(sequence_2_list, 1):
        fasta_file.write(f">sequence_{i}\n{seq}\n")

import csv

with open("random_rna_20.csv", 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["ID", "Sequence"])  # Header row
    for i, seq in enumerate(sequence_2_list, 1):
        writer.writerow([f"sequence_{i}", seq])
```
